# Оптимизация типов pandas

```python
df = pd.read_csv('/content/drive/MyDrive/python_data_analysis/Housing.csv')
df.head(20)
```

```python
df.describe()
```

```python
df.info(memory_usage="deep")
```

```python
df['furnishingstatus'].value_counts()
```

* price - судя по максимальной и минимальной величине, здесь подойдет int32
* area - аналогично, здесь достаточно int16
* bedrooms, bathrooms, stories, parking - числа небольшие, достаточно int8
* mainroad, guestroom, basement, hotwaterheating, airconditioning prefarea - содержат данные только yes/no, соответсвенно можно использовать категориальный тип
* furnishingstatus - здесь возможно только 3 варианта, категориальный тип также подходит


```python
df['price'] = df['price'].astype('int32')
df['area'] = df['area'].astype('int16')

for col_name in ['bedrooms', 'bathrooms', 'stories', 'parking']:
    df[col_name] = df[col_name].astype('int8')

for col_name in ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'prefarea', 'airconditioning', 'furnishingstatus']:
    df[col_name] = df[col_name].astype('category')
```

```python
df.info(memory_usage="deep")
```

Оптимизация позволила уменьшить занимаемую память с 251.7 KB до 10.8 KB, то есть на

```python
print(round((251.7 - 10.8)/251.7*100, 2),'%')
```
