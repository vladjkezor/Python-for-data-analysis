# Первая часть. Работа с API 'Ice and Fire'

```python
import pandas as pd
import requests
```

```python
base_url = "https://www.anapioficeandfire.com/api/"
```

```python
books_list = requests.get(f'{base_url}/books').json()
```

```python
books = pd.DataFrame(books_list)
books.head()
```

```python
houses_list = requests.get(f'{base_url}/houses').json()
```

```python
houses = pd.DataFrame(houses_list)
houses.head()
```

```python
houses_with_words_list  = requests.get(f'{base_url}/houses', params={'hasWords': True}).json()
```

```python
houses_with_words = pd.DataFrame(houses_with_words_list)
houses_with_words.head()
```
