Среднее количество выживших по классу

```python
data.groupby('Pclass')['Survived'].mean()*100
```

```python
data['FirstName'] = data['Name'].apply(lambda x: x.split('. ')[1])
```

Самое популярное имя

```python
data.groupby('Sex')['FirstName'].agg(lambda x: x.mode()[0])
```

Самое популярное имя на корабле в каждом классе

```python
data.groupby(['Sex', 'Pclass'])['FirstName'].agg(lambda x: x.mode()[0])
```

 Часть таблицы с пассажирами, возраст которых больше 44 лет

```python
data[data['Age']>44]
```

Часть таблицы с пассажирами, возраст которых меньше 44 лет мужского пола

```python
data[(data['Age'] < 44) & (data['Sex']=='male')]
```

количества n-местных кабин (в которых было 2, 3, 4, ... человека)

```python
data['Cabin'].dropna().value_counts().value_counts()
```

Всего n-местных кабин (в которых было 2, 3, 4, ... человека)
